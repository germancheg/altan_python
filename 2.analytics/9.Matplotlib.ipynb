- Диаграммы рассеивания

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.scatter(x, y)

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# подбор цвета
# colors = np.random.random(500)
# colors = np.linspace(0, 1, 500)
colors = x

# подбор размера точек
sizes = np.random.random(500) * 100

plt.scatter(x, y, c=colors, cmap="plasma", s=sizes)

plt.show()
```

```python
# данные по населению и по площади 
p = {
    "Якутск": 318768,
    "Нерюнгри": 57009,
    "Вилюйск": 11095
}
a = {
    "Якутск": 122,
    "Нерюнгри": 98.9,
    "Вилюйск": 15
}

# координаты (широта - lat и долгота - lon)
lat = {
    "Якутск": 62,
    "Нерюнгри": 56,
    "Вилюйск": 63
}
lon = {
    "Якутск": 129,
    "Нерюнгри": 124,
    "Вилюйск": 121
}

data = pd.DataFrame(
    {"population": p, "area": a, "latitude": lat, "longitude": lon}
)
data
```

```python
# визуализация данных

# логарифмирование данных по населению
pop_log = np.log10(data["population"])

plt.scatter(data["longitude"], data["latitude"], s=data["area"], c=pop_log)

# цветовая шкала
plt.colorbar(label="население (логарифм)" )

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# подбор цвета
# colors = np.random.random(500)
# colors = np.linspace(0, 1, 500)
colors = x

# подбор размера точек
sizes = np.random.random(500) * 100

symbol = 100

plt.scatter(x, y, c=colors, cmap="plasma", s=sizes, marker=f"${symbol}$")

plt.show()
```

```python
# объединение в одном окне графика и диаграммы рассеивания

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# модуляция
z = y + np.random.randn(1000) * 0.4

# диаграмма рассеиваний
plt.scatter(x, z, c=np.linspace(z.min(), z.max(), 1000), cmap="PiYG", alpha=0.5)

# график
plt.plot(x, y)

plt.show()
```
