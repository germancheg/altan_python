# Работа с Excel
- создание из датафрейма

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "A":{1:100, 2:200, 3:300},
    "B":{1:1000, 2:2000, 3:3000}
})
df
```

```python
# создание excel файла
df.to_excel("test.xlsx")
```

```python
# данные по Титанику
data = pd.read_csv("titanic.csv")
data.head()
```

```python
count_data = data.groupby(["survived", "pclass", "who"])["who"].count()
count_data
```

```python
mean_data = data.groupby(["survived", "pclass", "who"])["age"].mean()
mean_data
```

```python
df = pd.DataFrame({"count":count_data, "mean":mean_data})
df
```

```python
df.to_excel("titanic.xlsx")
```

```python
# создание excel файла с названием листа
df.to_excel("titanic.xlsx", sheet_name="Titanic")
```

```python
# создание excel файла с несколькими листами

# определение движка для записи нескольких датафреймов
excelWriter = pd.ExcelWriter("bigTitanic.xlsx", engine="xlsxwriter")

# создание excel
data.to_excel(excelWriter, sheet_name="full data") #первый датафрейм
df.to_excel(excelWriter, sheet_name="count and mean") #второй датафрейм

excelWriter.save()
```
