# Анализ данных из Росстат

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# загрузка данных
raw_data = pd.read_excel("cen-god.xls")
raw_data.head(10)
```

```python
# удаление верхней строки по индексу 
data_1 = raw_data.drop(0)
data_1.head()
```

```python
# удаление строк с пустыми (NaN) значениями
data_1 = data_1.dropna()
data_1.head()
```

```python
# строку с годами назначаем индексной строкой для столбцов
data_1.columns = data_1.loc[1, :]
data_1.columns
```

```python
data_1.head()
```

```python
data_1 = data_1.drop(1)
data_1.head()
```

```python
# приводим в порядок индексы столбцов: подправляем 2013 и приводим все значения в строковой тип 
col_arr = np.array(data_1.columns)
col_arr
```

```python
col_arr[col_arr == 2013] = 2013
data_1.columns = col_arr.astype("str")
data_1.columns
```

```python
# трансплонирование
data_2 = data_1.T
data_2.head()
```

```python
data_2.columns = data_2.iloc[0, :]
data_2 = data_2.drop(" ПОКАЗАТЕЛИ ")
data_2.head()
```

```python
# извлекаем года для временной шкалы по оси X, и попутно сконвертируем их в целочисленный тип
years = np.array(data_4.index).astype("int")
years
```

```python
data_4.columns
```

```python
def f1(data, key):
    return data.loc["1998":, key]
```

```python
# извлекаем данные по ценам
col_1 = f1(data_4, "Говядина (кроме бескостного мяса)")
col_2 = f1(data_4, "Дом отдыха, пансионат32), за день")
col_3 = f1(data_4, "Водка крепостью 40% об.спирта и выше9), за л")
col_4 = f1(data_4, "Электроэнергия в квартирах  без электроплит за минимальный объем потребления29), за 100 кВт·ч")
col_5 = f1(data_4, "Кинотеатры, за билет")
```

```python
years[years >= 1998]
```

```python
# визуализация данных

plt.figure(figsize=(12, 8))

plt.plot(years[years >= 1998], np.log10(col_5), label="Билеты в кино" )
plt.plot(years[years >= 1998], np.log10(col_4), label="Электроэнергия")
plt.plot(years[years >= 1998], np.log10(col_3), label="Водка")
plt.plot(years[years >= 1998], np.log10(col_2), label="Дом Отдыха")
plt.plot(years[years >= 1998], np.log10(col_1), label="Говядина")

plt.legend()
plt.xlabel("годы")
plt.ylabel("логарифмированные средние цены")
plt.title("Динамика средних цен")

plt.show()
```
